```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('mos2_soc_band.txt')

energy_data = data[:,1]

band_ene = np.reshape(energy_data,(32,120))

# max energy of valence band
vb_ene = max(band_ene[17,:]) +1.70580

# min emergy of conduction band
cb_ene = min(band_ene[18,:]) +1.70580

# band gap
band_gap = np.abs(cb_ene - vb_ene)

print(f"Band Gap of GaAs : {band_gap} ev ")

kpt= data[0:120,0]

# Plot
plt.figure(figsize=(8, 6))

for i in range(32):
    plt.plot(kpt, band_ene[i,:]+1.70580, 'b-')
    #plt.scatter(kpt, band_ene[i,:]+1.7789, marker='.', color = 'b')

plt.ylim(-4,4)
plt.fill_between(kpt, vb_ene, cb_ene, color='red', alpha=0.5, label=f'Band Gap = {band_gap} ev')
plt.xlabel("Kpoint distance")
plt.ylabel("Energy (eV)")
plt.title("Band Structure of MoS2 (with SOC)")
plt.legend(loc='upper left')
plt.grid(True, alpha = 0.3)
plt.show()
```
