```python
import numpy as np
import matplotlib.pyplot as plt
```

ENCUT convergence test

```python
# ENCUT values
encut = [150, 175, 200, 225, 250, 275, 300, 325, 350, 375, 400, 425, 450, 475, 500, 550, 600, 650, 700]

# E_0 values
e_0 = [-8.2631157, -8.2744978, -8.2788938, -8.280244, -8.2821785, -8.2828297, -8.2830863, -8.2831439, -8.2831792, -8.2832099, -8.2832933, -8.2833661, -8.2834427, -8.2835174, -8.283562, -8.2835871, -8.2836024, -8.2836357, -8.2836737]

# Plot
plt.figure(figsize=(8, 6))
plt.plot(encut, e_0, linestyle='-')
plt.scatter(encut, e_0, marker='o', color = 'b')

# Highlight the point where encut = 400
index_400 = encut.index(400)
plt.scatter(encut[index_400], e_0[index_400], color='r', label='Convergence Point (ENCUT = 400)')

plt.title('E_0 vs ENCUT for GaAs')
plt.xlabel('ENCUT')
plt.ylabel('E_0 (eV)')
plt.legend()
plt.grid(True, alpha = 0.5)
plt.show()
```
