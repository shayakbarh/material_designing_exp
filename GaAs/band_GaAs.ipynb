```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('band_GaAs.txt')

energy_data = data[:,1]

band_ene = np.reshape(energy_data,(16,240))

kpt= data[0:240,0]

# max energy of valence band
vb_ene = max(band_ene[3,:])

# min emergy of conduction band
cb_ene = min(band_ene[4,:])

# band gap
band_gap = np.abs(cb_ene - vb_ene)

print(f"Band Gap of GaAs : {band_gap} ev ")

# Plot the band structure
plt.figure(figsize=(8, 6))

for i in range(16):
    plt.plot(kpt, band_ene[i,:], 'b-')
    #plt.scatter(kpt, band_ene[i,:], marker='.', color = 'b')

plt.ylim([-3,3])
plt.fill_between(kpt, vb_ene, cb_ene, color='red', alpha=0.5, label=f'Band Gap = {band_gap} ev')
plt.xlabel("Kpoint distance")
plt.ylabel("Energy")
plt.title("Band Structure of GaAs")
plt.legend(loc='upper left')
plt.grid(True, alpha = 0.3)
plt.show()
```
